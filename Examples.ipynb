## Example 1: All cells have the same color.

```python
import numpy as np
import matplotlib.pyplot as plt
from voronoi_max_radius import plot_voronoi

points = np.array([[-.8, -.2], [0, .6], [0, 2], [1, 0], [1, 1], [1, 2],
                   [2.4, 0], [2, 1], [2, 2], [4, 4.5], [4, -.5]])
max_radius = 1.

xlim = (-3,6)
ylim = (-3,5)

fig, ax = plt.subplots(1,1)
plot_voronoi(ax,points, max_radius, xlim, ylim)
ax.set_aspect('equal')
plt.show()
```
